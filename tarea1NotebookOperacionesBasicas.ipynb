# Teorema de *Stokes*
Sea $\Sigma$ una superficie suave orientada en $\mathbb{R}^3$ con frontera $\partial{}\Sigma$. Si un campo vectorial $\bf{F}$ $= (P(x,y,z),Q(x,y,z),R(x,y,z))$ está definidio y tiene derivadas parciales continuas en una región abierta que contiene a $\Sigma$ entonces:

$\oint_{\partial{}\Sigma}\textbf{F}\cdot{d\textbf{r}}=\iint_{\Sigma}(\nabla{}\times{\textbf{F}})\cdot{d\textbf{S}}$

Nota: Los números amigos son aquellos donde la suma de sus divisores es igual al otro número (revisar wikipedia para ejemplos). 

Nota2: Recuerda que no puedes usar if, o crear funciones de más de 1 linea. Sí puedes crear funciones como f(x) = x^2. 

```julia
using LinearAlgebra
Divisores_propios = Int.([(i % j) == 0 for i in 1:1000, j in 1:1000] .* [(i/j) for i in 1:1000, j in 1:1000])  
#la primer matriz nos indica el lugar donde hay divisores propios y la segunda el valor de estos (se sugiere probar cada matriz por separado para ver mejor lo que contiene c/u)
```

```julia
Divisores_propios1 = Divisores_propios[2:1000,2:1000]  #le quitamos la primer columna y primer renglón para hacer más fáciles los cálculos
```

```julia
suma_DP = [sum(Divisores_propios1[i,:]) for i in 1:999]    #La suma de los divisores propios desde el 2 hasta el 1000
```

```julia
lista1 = [[(i+1) , suma_DP[i]] for i in 1:999]    #Duplas del número y la suma de sus divisores propios
```

```julia
lista2 = [[suma_DP[i] , (i+1)] for i in 1:999]   #Las mismas duplas pero en el orden contrario
```

```julia
num_am = [lista1[i] == lista2[j] for i in 1:999, j in 1:999]    #Se buscan duplas iguales (lo que representaría que encontramos numeros amigos)
```

```julia
A1 = [ Int.(zeros(1,999)) ;(num_am)]
Matriz_final = [Int.([1;zeros(999,1)]) A1]                    #por último se añaden la primera columna y el primer renglón (sabemos que el 1 es solamente amigo de sí mismo) 
```

Ahora veamos quiénes son números amigos

```julia
findall(isodd,Matriz_final)   #Buscamos los numeros impares en la matriz (solo hay 0s y 1s)
```
