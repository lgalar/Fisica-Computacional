```julia
function coef(l,n)
    """Esta función tiene como argumentos una lista y un entero. Nos arroja los valores de las constantes 
       de la lista l"""
    if length(l)-1 == n 
        A = [BigFloat(i^j) for j ∈ 0:n, i ∈ l]
        B = zeros(n+1)
        B[n+1] = 1
        return c = inv(A)*B
    else 
        return -1
    end
end
```

```julia
coef([1,1.5,2,2.5,3],4)
```

```julia
function fun(l, n, f, x; Δx = 0.0001)
    c = coef(l,n)
    return [sum(c[i]*factorial(n)*f(j+l[i]*(Δx))/((Δx)^n) for i ∈ 1:length(l)) for j ∈ x]
end
```

```julia
struct D{T1 <: Number, T2 <: Number} <: Number
    valor::T1
    derivada::T2
end
```

```julia
# definimos cómo operar con los el número dual D, +,-,*,/,^
import Base.+, Base.-, Base.*, Base./, Base.^
+(a::D, b::D) = D(a.valor + b.valor, a.derivada + b.derivada)
+(a::D, b::Number) = D(a.valor + b, a.derivada)
+(b::Number, a::D) = D(a.valor + b, a.derivada)
-(a::D, b::D) = D(a.valor - b.valor, a.derivada - b.derivada)
-(a::D, b::Number) = D(a.valor - b, a.derivada)
-(b::Number, a::D) = D(-a.valor + b, -a.derivada)
*(a::D, b::D) = D(a.valor*b.valor, a.derivada*b.valor+b.derivada*a.valor)
*(a::D, b::Number) = D(b*a.valor,b*a.derivada)
*(b::Number, a::D) = D(b*a.valor,b*a.derivada)
/(a::D, b::D) = D(a.valor/b.valor,(a.derivada*b.valor-a.valor*b.derivada)/b.valor^2)
/(a::D, b::Number) = D(a.valor/b,a.derivada/b)
/(b::Number, a::D) = D(b/a.valor,-b*a.derivada/a.valor^2)
^(a::D, b::Real) = D(a.valor^b, b*a.valor^(b-1)*a.derivada)
^(a::D, b::Int) = D(a.valor^b, b*a.valor^(b-1)*a.derivada)
```

```julia
derivada(f, x::D) = f(D(x, D(1,0))).derivada

function derivadan(f, x, n) 
    if n == 0
        return f(x)
    end
    g(x) = derivada(f, x)   # se define una nueva función a derivar usando la derivada de la función anterior. 
    x = derivadan(g, x, n-1)
end

function derivadasn(f, x, n) 
    if n == 0
        return f(x)
    end
    g = "derivada(x->f(x), x)"  # alternativamente, para no usar la recursividad, se puede utilizar la meta-programación. 
    for i in 1:n-1
        g = "derivada(x->$g, x)"  # sin embargo, es más lento, porque la función eval es lenta. Pero básicamente esto es lo
    end                           # sucede con la versión iterativa. 
    eval(Meta.parse(g))
end
```

```julia
using Plots, Random, LaTeXStrings
function nosimbolica(x)
    Random.seed!(Int(abs((trunc(x)))))
    a=rand([-1,1])
    return a*((abs(x)-trunc(abs(x)))^4/4-(abs(x)-trunc(abs(x)))^3/2+(abs(x)-trunc(abs(x)))^2/4)
end

# Cada ciclo de la funcion es aleatorio, por lo que no se puede derivar simbolica ni automaticamente.

d(f, t; Δt = 0.0001) = (f(t+Δt)-f(t))/Δt 

f1(x) = d(nosimbolica,x, Δt = big(1e-6))
f2(x) = d(f1,x, Δt = big(1e-6))
f3(x) = d(f2,x, Δt = big(1e-6))
f4(x) = d(f3,x, Δt = big(1e-6))

x = BigFloat(-15):BigFloat(0.01):15

plot(x, nosimbolica.(x), label = L"f",ylim=[-.05,.05])




```
